```python
sns.histplot(df["age"])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df["bmi"])
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df["children"])
plt.title("Children Distribution")
plt.xlabel("Children")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df["expenses"])
plt.title("Expenses Distribution")
plt.xlabel("Expenses")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x=df["sex"])
plt.title("Count of Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x=df["smoker"])
plt.title("Count of Smoker")
plt.xlabel("Smoker")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x=df["region"])
plt.title("Count of Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.show()
```

```python
corelated_data = df.corr(numeric_only=True)
sns.heatmap(corelated_data,annot=True)
```
